# Question 1 

### Could audio be convereted to pictures ? 

#### Answer yes or no  , then  explain briefly to support your choice . 


Yes, audio can be converted to images and it is a common technique. A common image to convert to is a spectrogram, which is a graphical representation of audio frequency strength. Audio can also be converted to other kinds of graphs such as waveform graphs. 

# Question 2 

### Explain briefly the role of CNN architecture in Audio deep learning model .

CNNs are used to process audio via a transformed state; spectrograms. Audio is converted to spectrograms (most commonly, other graphical formats are certainly possible) and the CNN can then work with the resulting image. Since the values in locations around a certain data point within an audio file define how humans perceive sound, CNNs are a good architecture to use as they can pick up the local relationships in the data and the variety of filters available allow a large number of attributes in the data to be detected.

## Task 1

### use above library and load the wav file


```python
audio, sr = librosa.load(file)
print(audio)
print(sr)
```


## Task 2

### Once the wav file is loaded , plot the waveform for the audio file and display it 
[Hint : X axis (Time in seconds) , Y axis (Amplitude)]

```python
# Values
plt.plot(audio)

# Labels
plt.title("Waveform")
plt.xlabel("Time (Sec.)")
plt.ylabel("Amplitude")

# xticks
a = audio.shape[0]
ss = a / ((a / sr) * 2)
xticks = np.arange(0, a + 1, ss)
xtick_labels = np.arange(0, a / sr + 0.5, 0.5)
plt.xticks(xticks, xtick_labels)

plt.show()
```
