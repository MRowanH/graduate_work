```python
from bs4 import BeautifulSoup
import os
import re
import requests

text = None
with open("source.txt", "r") as file:
    text = file.read()

links = []
soup = BeautifulSoup(text, "html.parser")
for link in soup.findAll("a"):
    href = link.get("href")
    if "GSE" in href:
        links.append(href)
```

```python
root = "https://www.ncbi.nlm.nih.gov"
counts = []
length = len(links)
for i, l in enumerate(links):
    print(i, "/", length)
    url = "%s%s"%(root, l)
    r = requests.get(url)
    text = r.content.decode("utf-8")
    c = re.findall("(Samples \([0-9]+\))+", text)[0]
    c = re.findall("[0-9]+", c)[0]
    counts.append(int(c))
counts = sorted(counts, reverse = True)
```
