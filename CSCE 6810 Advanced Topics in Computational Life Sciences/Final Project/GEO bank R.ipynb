# Function

```r
get_gse <- function(gse) {
    paste(gse, " starting...")

    tryCatch(
        expr = {
        
            # Code from Danyang start
            gset = AnnoProbe::geoChina(gse)

            gset = gset[[1]]
            phenoDat <- pData(gset) 

            group_list = factor(c(rep('Normal',60), rep('SLE',1760)))

            group_list = ifelse(str_detect(phenoDat$title,"SLE"),"SLE","normal")

            group_list = factor(group_list,
                                levels = c("normal","SLE")
                                )

            exprSet=exprs(gset)

            exprSet2 = normalizeBetweenArrays(exprSet)

            ex <- exprSet
            qx <- as.numeric(quantile(ex, c(0., 0.25, 0.5, 0.75, 0.99, 1.0), na.rm=T))
            LogC <- (qx[5] > 100) ||
                (qx[6]-qx[1] > 50 && qx[2] > 0) ||
                (qx[2] > 0 && qx[2] < 1 && qx[4] > 1 && qx[4] < 2)

            if (LogC) { 
                ex[which(ex <= 0)] <- NaN
                exprSet <- log2(ex)
                }

            (gpl=gset@annotation)
            ids <- idmap(gpl, type = 'soft')

            exprSet <- filterEM(exprSet,ids)
            # Code from Danyang end

            }, 
        warning = function(w) {
            print(w)
            }, 
        error = function(e) {
            print(e)
            }, 
        finally = function(f) {
            print(f)
            }
    )
    
    paste(gse, " finished")
}
```
