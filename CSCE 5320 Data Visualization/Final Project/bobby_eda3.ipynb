```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import openpyxl
import xlrd

df = pd.read_excel(r'C:\Users\15126\OneDrive\csce5320\final_project\data\Building.xls')
print(df.head())
print(df.info())
```

```python
print(df.describe())
```

```python
print(df.columns)
```

```python

for col in df.select_dtypes(include=[np.number]).columns:
    df[col].fillna(df[col].median(), inplace=True)


for col in df.select_dtypes(include=['object']).columns:
    df[col].fillna('Unknown', inplace=True)


print(df.describe(include='all'))


sns.countplot(data=df)
plt.title('Distribution of Accident Severity')
plt.xticks(rotation=45) 
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='SectorName', data=df)
plt.title('Distribution of Sector Names')
plt.xticks(rotation=45)
plt.show()
```

```python
#create a scatterplot of the geographical distribution of properties
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Longitude', y='Latitude', hue='Region', data=df)
plt.title('Geographical Distribution of Properties')
plt.show()
```
