```python
#Importing the correct libraries

import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
#Reading in the dataset

#car_heat = pd.read_csv('price_paid_records.csv')
car_heat = pd.read_csv('Road Accident Data.csv')
```

```python
#Removing rows with null values

car_heat = car_heat.dropna()
```

```python
#Creating the heat map comparing price and district

plt.figure(figsize=(10,6))
plt.hexbin(car_heat['Latitude'], car_heat['Longitude'], gridsize=50, cmap='inferno')
plt.colorbar(label='count')
plt.title('Heatmap Latitude and Longitude for Crashes')
plt.xlabel('Latitude')
plt.ylabel('Longitude')
plt.show()
```

```python
#Reading in the dataset

house_hist = pd.read_csv('price_paid_records.csv')
```

```python
#Removing null values

house_hist = house_hist.dropna()
```

```python
#Isolating the district variable for analysis and showing the top 10 highest concentrations of houses

count = house_hist['District'].value_counts()
top10_count =count.head(10)
```

```python
#Creating the histogram

plt.figure(figsize=(10,6))
top10_count.plot(kind='bar', color='green')
plt.title('Analysis of Houses purchased by District')
plt.xlabel('District')
plt.ylabel('Number of Houses')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
#Creating new variables to create bar plot showing the top 10 highest prices by district

car_x = 'Local_Authority_(District)'
car_y = 'Number_of_Casualties'
avg_crashes = car_heat.groupby(car_x)[car_y].mean()
top10 = avg_crashes.nlargest(10)
filtered_top10 = car_heat[car_heat[car_x].isin(top10.index)]
```

```python
#Creating the bar chart

plt.figure(figsize=(10,6))
filtered_top10.groupby(car_x)[car_y].mean().plot(kind='bar', color='blue')
plt.title('Number of Car Casualties by District')
plt.xlabel('District')
plt.ylabel('Number of Casualties')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
