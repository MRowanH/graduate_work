```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv(r'C:\Users\15126\OneDrive\csce5320\final_project\data\price_paid_records.csv')
print(df.head())
print(df.info())
```

```python
print(df.columns)
```

```python
print(df.describe())
```

```python
#sort and show the distribution of property types

sns.countplot(x='Property Type', data=df)
plt.title('Distribution of Property Types')
plt.xticks(rotation=45)
plt.show()
```

```python
#group by median price and then select the top 20 cities to visulaize
town_prices = df.groupby('Town/City')['Price'].median()


top_town_prices = town_prices.sort_values(ascending=False).head(20)


top_town_prices.plot(kind='bar', color='skyblue')  
plt.title('Top 20 Cities by Median Property Prices')
plt.xlabel('Town/City')
plt.ylabel('Median Price')
plt.xticks(rotation=90) 
plt.show()
```
