## Kaggle

### General Preprocessing

```python
def convert_senses(seq):
    converter = {"A": "U",
                 "C": "G",
                 "G": "C",
                 "U": "A"
                 }
    out = ""
    for c in seq:
        out = converter[c] + out
    return out
```

```python
# Add generated sense/antisense sequences
cols = list(kaggle_df.columns)
idx = cols.index("Sequence")
df_new = []
for i, row in kaggle_df.iterrows():

    # Get converted sequence
    seq_new = convert_senses(row["Sequence"])

    # Add both rows to the new dataframe
    for j in range(2):
        df_new.append([])
        for c in cols:
            df_new[-1].append(row[c])

    # Add converted sequence to dataframe
    df_new[-1][idx] = seq_new

    # Add sense/antisense label to rows
    df_new[-2].append("sense")
    df_new[-1].append("antisense")

    # if i == 4:
    #     break

# Convert matrix to dataframe
cols.append("Sense/Antisense")
kaggle_df = pd.DataFrame(df_new, columns=cols)
kaggle_df
```

```python
# Get G and U count for first 1/2 of sequence
for c in ["A", "C", "G", "U"]:
    kaggle_df[c] = kaggle_df["Sequence"].apply(lambda x: float(len([x[i] for i in range(int(len(x)/2)) if x[i] == c])))
    kaggle_df[c] = kaggle_df[c].astype(np.int64)
    if c not in cols:
        cols.append(c)
kaggle_df
```

```python
# Normalize columns
for c in cols:

    # Work only with numerical columns that are not Start or End
    if (kaggle_df[c].dtype == np.int64 or kaggle_df[c].dtype == np.float64) and (c != "Start" and c != "End"):
        print(c)

        # Round Activity column
        if c == "Activity":
            kaggle_df[c] = kaggle_df[c].apply(lambda x: round(x, 0))

        # Get min and max
            mi = 0
            ma = 1
        else:
            mi = kaggle_df[c].min()
            ma = kaggle_df[c].max()
        print("\t", mi, "\n\t", ma)

        # Define normalization function
        norm = lambda x: (x-mi)/(ma-mi)

        # Normalize the column
        kaggle_df[c] = kaggle_df[c].apply(norm)

kaggle_df
```

### Dataset Subsets

```python
def subset_dataframe_and_save(df, df_name):
    save_dir = data_dir + "kaggle_df_" + df_name + "_"

    # Calculate antisense
    kaggle_df_one = kaggle_df.loc[kaggle_df["Sense/Antisense"] == "sense"]
    kaggle_df_both = kaggle_df
    save_df_to_csv(save_dir + "one.csv", kaggle_df_one)
    save_df_to_csv(save_dir + "both.csv", kaggle_df_both)

    # Sequences only
    kaggle_df_seqs = kaggle_df_one[["Target seq", "Start", "End", "Sequence", "Target gene", "Sense/Antisense"]]
    save_df_to_csv(save_dir + "seqs.csv", kaggle_df_seqs)

    # Keep activity above limit
    lim = 0.7
    kaggle_df_activity = kaggle_df_one.loc[kaggle_df_one["Activity"] >= lim]
    kaggle_df_activity = kaggle_df_activity[["Target seq", "Start", "End", "Sequence", "Target gene", "Sense/Antisense"]]
    save_df_to_csv(save_dir + "activity.csv", kaggle_df_activity)

    # Keep some numeric features
    keeps = ["Target seq", "Start", "End", "Sequence", "Target gene", "Sense/Antisense",
            "Dif_5-3", "A", "C", "G", "U", "Cons_Sum",
            ]
    kaggle_df_nums = kaggle_df_one[keeps]
    save_df_to_csv(save_dir + "nums.csv", kaggle_df_nums)
```

```python
# 80/20
subset_dataframe_and_save(kaggle_df_80_train, "80_train")
subset_dataframe_and_save(kaggle_df_80_test, "80_test")

# 90/10
subset_dataframe_and_save(kaggle_df_90_train, "90_train")
subset_dataframe_and_save(kaggle_df_90_test, "90_test")
```
