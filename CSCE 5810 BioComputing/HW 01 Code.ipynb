## 1. (20 pts) Implement problem 1N (d-neighborhood)

```python
# Get the hamming distance between 2 strings
# Assume lenghts are equal
def hamming(txt1, txt2):
    dis = 0  # Start with 0 distance
    for i in range(len(txt1)):  # Iterate over all positions

        # If characters at position i are not identical, increase the count
        if txt1[i] != txt2[i]:
            dis += 1

    return dis

# Generate all of the neighbors of txt with a hamming distance less than or equal to d
def neighbors(txt, d):

    # If d is 0, then there are no neighbors and the only string returned is the original
    if d == 0:
        return txt

    # If the lenght of txt is 1, then the neighborhood consists of the nucleotides
    elif len(txt) == 1:
        return ["A", "C", "G", "T"]

    # Otherwise, build the neighborhood
    else:
        neighborhood = set()  # Use a set to avoid identical strings
        suffix_n = neighbors(txt[1:], d)  # Get the neighborhood of the suffix of txt

        # For each string in the suffix's neighborhood, generate 4 new strings each with a different nucleotide in front
        for s in suffix_n:  # Iterate over the suffix's neighborhood
            for n in ["A", "C", "G", "T"]:  # Iterate over the nucleotides
                new = n + s  # Get new string

                # Keep the new string if the hamming distance does not exceed d
                if hamming(new, txt) <= d:
                    neighborhood.add(new)

        # Return the neighborhood of txt
        return list(neighborhood)  # Converted to a list for iteratability

# Display all neighbors
for k_m in neighbors("ACG", 2):
    print(k_m)
```

## 2. (20 pts) Implement 1J, Frequent Words with Mismatches and Reverse Complements (by adapting the approach on page 48 of the textbook, 1.15 on Stepik.org)

```python
# Get the reverse compliment of txt
def reverse_compliment(txt):
    converter = {"A": "T",  # Converter dictionary for ease of use
                 "C": "G",
                 "G": "C",
                 "T": "A",
                 }
    out = ""  # Output string initilized to an empty string

    # For each character in txt, append it's compliment to the beginning of the output string
    for c in txt:
        out = converter[c] + out

    return out

# Get all possible k-mer combinations for size k
def gen_combos(k, k_mers=None):
    nucs = ["A", "T", "C", "G"]  # List of nucleotides

    # If k_mers is none, set the initail k-mers list to the list of nucleotides and make a recursive call
    if k_mers is None:
        return gen_combos(k-1, nucs)

    # Otherwise build on the k_mers list
    else:
        k_mers_new = []  # Start a new k-mers list
        for k_m in k_mers:  # Itearte over the old k-mers list
            for n in nucs:  # Itearte over the nucleotides

                # For each of the old k-mers, geneate 4 new k-mers with a different nucleotide at the end of each
                k_mers_new.append(k_m + n)

        # If k cannot be decremented, return the new list of k-mers
        if k <= 1:
            return k_mers_new

        # Otherwise, make a recursive call
        else:
            return gen_combos(k-1, k_mers_new)

# Get the counts of each k-mer occurance in a string
def get_kmer_counts(txt, k):
    k_mers = {}  # Inialzie an empty dictionary
    for i in range(len(txt) - k + 1):  # Iterate over each k-mer in txt
        km = txt[i:i+k]  # Get the current k-mer

        # Add the k-mer to the dictionary if necessary
        if km not in k_mers:
            k_mers.update({km: 0})

        # Update the k-mer's count
        k_mers[km] += 1

    return k_mers

# Get the most fequent k-mers with reverse compliments and mismatches
def get_freq_words_mismatch_rev_comp(txt, k, d):
    k_mers = get_kmer_counts(txt, k)  # Get the counts of the k-mers in txt
    k_cts = gen_combos(k)  # Gereate all possible k-mers of size k
    k_cts = dict(zip(k_cts, [0 for i in range(len(k_cts))]))  # Convert list to dictionary with values of 0

    c_max = 0  # Initilize max count tracker to 0

    for km in k_mers:  # Iterate over ever k-mer in txt
        rev = reverse_compliment(km)  # Get the reverse compliment of the current k-mer
        for kc in k_cts:  # Iterate over every possible k-mer

            # If the hamming distance between the k-mers does not exceed d, add the count to the geneated dictionary
            if hamming(km, kc) <= d:
                k_cts[kc] += k_mers[km]

            # If the hamming distance between the reverse compliment and the k-mer does not exceed d, add the count to the geneated dictionary
            if hamming(rev, kc) <= d:
                k_cts[kc] += k_mers[km]

            # Check if the current count is larger than the max, set it to the new max if so
            if k_cts[kc] > c_max:
                c_max = k_cts[kc]


    freqs = []  # Initilize the frequent k-mers list
    for kc in k_cts:  # Itreate over every possible k-mer

        # If the count of the k-mer is equal to the max, add it to the frequents list
        if k_cts[kc] == c_max:
            freqs.append(kc)

    return freqs

txt = "ACGTTGCATGTCGCATGATGCATGAGAGCT"
k = 4
d = 1
get_freq_words_mismatch_rev_comp(txt, k, d)
```

```python
# Get the list of the running skew total for a gene sequence
def get_skew(txt):

    # Initilize counters to 0 and skew to an empty list
    c_c = 0
    c_g = 0
    skew = []

    # Iterate over every nucleotide in txt
    for i, n in enumerate(txt):

        # If the nucleotide is a C or G, increment the counters
        if n == "C":
            c_c += 1
        elif n == "G":
            c_g += 1

        # Append the running G - C total to the skew list
        skew.append(c_g - c_c)

    return skew

# Get the skew of txt
skew = get_skew(txt)

# Shift genome by 1/4 to get a more friendly skew plot
pt = int(len(txt)/4)
txt_shift = txt[pt:] + txt[:pt]

# Get the skew of the shifted txt
skew_shift = get_skew(txt_shift)
```

```python
# Get indices of where motifs occur
def locate_motifs(txt, results, oriC_start, oriC_end):
    sig = {}  # Initilize empte dictionary to track how often an motif is inside/outside of the oriC region

    # Itrate through motifs and their d values
    for d in results:
        sig.update({d: {}})
        for k in results[d]:
            for b in results[d][k]:

                sig[d].update({b: {"in": 0, "out": 0}})  # Add the current box to the dictionary
                for i in range(len(txt) - len(b) + 1):  # Iterate through txt

                    # If a match for the motif is found, increment the counter for an index being inside/outside the oriC region
                    if hamming(txt[i:i+len(b)], b) <= d:
                        if i >= oriC_start and i+len(b) < oriC_end:
                            sig[d][b]["in"] += 1
                        else:
                            sig[d][b]["out"] += 1

                # Calculate the ratio for how often a motif is inside of the oriC region
                sig[d][b].update({"ratio": sig[d][b]["in"]/(sig[d][b]["in"] + sig[d][b]["out"])})

    return sig

# Display where potential DnaA boxes occur
def print_dnaa_box_stats(txt, results, oriC_start, oriC_end):

    # Get the location of motifs
    sig = locate_motifs(txt, results, oriC_start, oriC_end)

    # Calculate baseline
    base = 500 / len(txt)

    # Build table out of dictionary
    table = []
    for d in sig:
        for b in sig[d]:
            table.append([d, b])
            for k in sig[d][b]:
                table[-1].append(sig[d][b][k])

            # Check if appearance ratio is over baseline and add to table
            if sig[d][b]["ratio"] > base:
                table[-1].append("yes - %f"%(sig[d][b]["ratio"] - base))
            else:
                table[-1].append("no")

    # Display table
    print(tabulate(table, headers=["d", "Box", "In", "Out", "Ratio", "Over Baseline"], tablefmt="grid"))

# Display baseline statistic
for k in [6, 9]:
    print("%d-mer baseline ratio is %f"%(k, 1/4**k))
print("The oriC region represents %f%% of the DNA sequence"%(500/len(txt)))

# Display for txt
print("\nNon-Shifted Genome:")
print_dnaa_box_stats(txt, results, 0, 500)

# Display for txt_shifted
print("\nShifted Genome:")
print_dnaa_box_stats(txt_shift, results_shift, mi, ma)
```
