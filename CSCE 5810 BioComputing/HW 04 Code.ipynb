# Function

```python
def plot_scores(r_dict, key):

    # Get x and y
    for p in PAIRS:
        x = []
        y = []
        for m in r_dict[p]:
            x.append(m)
            y.append(r_dict[p][m])
        plt.plot(x, y, label=p)

    # Legend
    plt.legend(loc="center right")

    # Label plot
    plt.xlabel("%s Matrix"%(key))
    plt.ylabel("Score")
    plt.title("Trend of %s Matricies"%(key))

    plt.show()
```

# Plots

```python
plot_scores(results["blosum"], "BLOSUM")
```

```python
plot_scores(results["pam"], "PAM")
```
